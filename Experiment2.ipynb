```python
# to show that number is Armstrong number or not
def count(num):
    count = 0
    while(num / 10 != 0):
        count  = count + 1
        num = num // 10
    return count

def check_Armstring(num,count):
    sum = 0
    while(num / 10 !=0):
        n = num % 10
        sum = sum + (n ** count)
        num = num // 10
    return sum

# main
num = int(input())
count_digit = count(num)

check_no = check_Armstring(num,count_digit)

if(check_no == num):
    print("it is Armstring number")
else:
    print("it is not a Armstring number")
```

```python
#find factorial by recursion method
def find_factorial(num):
    factorial = 1
    if(num ==1):
        return factorial
    else:
        return(num * find_factorial(num-1)) 


num = int(input())
print("The factorial of",num,"is",find_factorial(num))
```

```python
# Python program to find the HCF of two input number

def find_hcf(n1, n2):
    while(n1>=n2):
            rem = n1%n2
            if(rem != 0):
                n2 = rem
                n1 = n2
            else:
                return n2
    while(n2>=n1):
            rem = n2%n1
            if(rem != 0):
                n1 = rem
                n2 = n1
            else:
                return n1
        

n1=int(input())
n2 = int(input())
print(find_hcf(n1,n2))
```

```python
# Python program to find the L.C.M of two input number

def find_hcf(n1, n2):
    while(n1>=n2):
            rem = n1%n2
            if(rem != 0):
                n2 = rem
                n1 = n2
            else:
                return n2
    while(n2>=n1):
            rem = n2%n1
            if(rem != 0):
                n1 = rem
                n2 = n1
            else:
                return n1

def find_lcm(n1,n2):
    hcf=find_hcf(n1,n2)
    lcm = (n1*n2)//hcf
    return lcm

n1=int(input())
n2 = int(input())
print(find_lcm(n1,n2))
```

```python
# To find the area of circle, square, rectangle, triangle and sphere

import math,sys
def Area_triangle(a, b,c):
    s = (a+b+c)/2
    # calculation of the area of triangle

    area = math.sqrt((s*(s-a)*(s-b)*(s-c)))
    print("Area of triangle is:  ",area)

def Area_circle(radius):
    area = math.pi*(radius * radius)
    print("Area of circle: ", area)

def Area_Square(side):
    area = side * side
    print("Area of square: ", area)

def Area_Rectangle(length,beadth):
    area= length * beadth
    print("Area of Rectangle: ",area)

def Area_sphere(radius):
    area = 4*math.pi*(radius*radius)
    print("Area of sphere: ",area)
    

#main
print("Area of 1. Triangle, 2. Circle, 3. Square, 4. Rectangle, 5. sphere, 6.exit")
while(True):
    
    option = int(input())
    if(option == 1):
        print("Enter the length of three side of Triangle")
        a=int(input())
        b = int(input())
        c = int(input())
        Area_triangle(a, b, c)
    elif(option == 2):
        print("Enter the length of radius of circle")
        a=int(input())
        Area_circle(a)
    elif(option == 3):
        print("Enter the side of the square")
        a = int(input())
        Area_Square(a)
    elif(option == 4):
        print("Enter the length and breadth of rectangle")
        a = int(input())
        b = int(input())
        Area_Rectangle(a, b)

    elif(option == 5):
        print("Enter the radius of sphere")
        a = int(input())
        Area_sphere(a)
    elif(option == 6):
        print("Exit")
        break
    else:
        print("worng input retry")
        continue
        
               
            
            
            
            
            
```
